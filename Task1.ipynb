```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,11)
y = np.arange(0,11)
#y = np.random.randint(10,100, size=11)
plt.figure(figsize=(6,4))
plt.plot(x,y,marker='^',linestyle='-',color='purple')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot")
plt.grid(True)
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [40, 70, 30, 90]
plt.figure(figsize=(4,4))
plt.bar(products,sales,color='orange')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Vertical Bar Plot")
plt.show()
```

```python
fig, axis = plt.subplots(2,2,figsize=(4,4))
axis[0, 0].plot(x, y)
axis[0, 0].set_title("Line Plot")
axis[0, 1].bar(products, sales)
axis[0, 1].set_title("Bar Plot")
axis[1, 0].scatter(x, y)
axis[1, 0].set_title("Scatter Plot")
axis[1, 1].barh(products, sales)
axis[1, 1].set_title("Horizontal bar Plot")
```

```python
data = np.random.randn(200)
plt.hist(data, bins=20, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Categories")
plt.ylabel("Value")
plt.show()
```
