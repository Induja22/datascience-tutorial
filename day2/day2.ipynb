```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(38)
data = {
    "x" : np.arange(1,10),
    "y" : np.random.randint(15,60,size=9)
}
df = pd.DataFrame(data)
print(df)
plt.figure(figsize = (5,5))
plt.plot(df["x"],df["y"],marker='.',color='purple')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = np.random.randint(10,100, size=10)
data = np.random.randn(100)
print(x)
print(y)
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,y,marker='x',linestyle='--',color='purple')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='purple')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(x,y,color='orange')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(x,y,color='red')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Horizontal Bar Plot")
plt.show()
```
